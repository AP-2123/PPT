**Question-1**

You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.
```
Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7
```

````python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def inorderTraversal(root, result):
    if root:
        inorderTraversal(root.left, result)
        result.append(root.val)
        inorderTraversal(root.right, result)

def convertToBST(root, sorted_list):
    if root:
        convertToBST(root.left, sorted_list)
        root.val = sorted_list.pop(0)
        convertToBST(root.right, sorted_list)
````

````python
# Example tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Perform in-order traversal to get sorted elements
inorder_result = []
inorderTraversal(root, inorder_result)

# Sort the elements
sorted_elements = sorted(inorder_result)

# Convert binary tree to binary search tree
convertToBST(root, sorted_elements)

# Helper function to print the binary tree in in-order traversal
def printInorder(root):
    if root:
        printInorder(root.left)
        print(root.val, end=" ")
        printInorder(root.right)

# Print the resulting binary search tree
print("Binary Search Tree (in-order traversal):")
printInorder(root)
````

**Input-1:**

node-1 = 6

node-2 = 14

**Output-1:**

The distance between the two keys = 4

**Input-2:**

node-1 = 3

**Output-2:**

The distance between the two keys = 2

````python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return TreeNode(value)
    
    if value < root.val:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    
    return root

def findDistance(root, node1, node2):
    if root is None:
        return 0

    if node1.val < root.val and node2.val < root.val:
        return findDistance(root.left, node1, node2)
    elif node1.val > root.val and node2.val > root.val:
        return findDistance(root.right, node1, node2)
    else:
        return distanceFromRoot(root, node1) + distanceFromRoot(root, node2)

def distanceFromRoot(root, node):
    if root.val == node.val:
        return 0
    elif node.val < root.val:
        return 1 + distanceFromRoot(root.left, node)
    else:
        return 1 + distanceFromRoot(root.right, node)

def findNode(root, target):
    if root is None or root.val == target:
        return root

    if target < root.val:
        return findNode(root.left, target)
    else:
        return findNode(root.right, target)
````

````python
# Example values and nodes
values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
node1_value = 6
node2_value = 14

# Create the Binary Search Tree
root = None
for value in values:
    root = insert(root, value)

# Find the nodes corresponding to node1 and node2
node1 = findNode(root, node1_value)
node2 = findNode(root, node2_value)

# Calculate the distance between node1 and node2
distance = findDistance(root, node1, node2)

# Print the distance
print("The distance between the two keys =", distance)
````

**Question-3:**

Write a program to convert a binary tree to a doubly linked list.
```
Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35
```

````python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.prev = None
        self.next = None

def convertToDoublyLinkedList(root):
    if root is None:
        return None

    head, _ = convertToDoublyLinkedListHelper(root)
    
    # Find the tail of the doubly linked list
    tail = head
    while tail.next:
        tail = tail.next
    
    # Set the prev pointer of the head to None
    head.prev = None
    
    # Set the next pointer of the tail to None
    tail.next = None
    
    return head

def convertToDoublyLinkedListHelper(node):
    if node is None:
        return None, None
    
    # Convert left subtree recursively
    if node.left:
        left_head, left_tail = convertToDoublyLinkedListHelper(node.left)
        
        # Set the next pointer of the rightmost node to the current node
        left_tail.next = node
        node.prev = left_tail
    else:
        left_head = node
    
    # Convert right subtree recursively
    if node.right:
        right_head, right_tail = convertToDoublyLinkedListHelper(node.right)
        
        # Set the prev pointer of the leftmost node to the current node
        right_head.prev = node
        node.next = right_head
    else:
        right_tail = node
    
    return left_head, right_tail

def printDoublyLinkedList(head):
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.next
    print()

# Example tree
root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

# Convert the binary tree to a doubly linked list
head = convertToDoublyLinkedList(root)

# Print the doubly linked list
print("Output:")
printDoublyLinkedList(head)
````

````python
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.nextRight = None

def connectNodesAtSameLevel(root):
    if not root:
        return

    # Create a queue for level order traversal
    queue = [root]

    while queue:
        # Get the size of the current level
        level_size = len(queue)

        # Process all nodes in the current level
        for i in range(level_size):
            # Dequeue the front node
            node = queue.pop(0)

            # Set the nextRight pointer of the current node
            if i < level_size - 1:
                node.nextRight = queue[0]

            # Enqueue the left and right children of the current node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

def printLevelOrderTraversal(root):
    if not root:
        return

    node = root
    while node:
        current_level = node
        while current_level:
            print(current_level.val, end=" -> ")
            current_level = current_level.nextRight
        print("-1")
        node = node.left

# Example tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connectNodesAtSameLevel(root)

# Print level order traversal with nextRight pointers
print("Output:")
printLevelOrderTraversal(root)
````
